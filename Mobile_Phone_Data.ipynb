```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/best-selling-mobile-phones.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10, 3))
# plt.xtick(rotation = 'rotation')
sns.countplot(x= df['form'], order=df.form.value_counts().sort_values(ascending = False).index)
plt.title('Most sold mobile phone form')
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x=df['smartphone'])
plt.title('The kind of phone that sold the most')
```

```python
plt.figure(figsize=(10, 5))
plt.xticks(rotation = 'vertical')
sns.countplot(x=df['manufacturer'], order=df.manufacturer.value_counts().sort_values
              (ascending = True).index)
plt.title('Manufacturer that sold the most')
plt.xticks()
```

```python
plt.figure(figsize=(10, 5))
plt.xticks(fontsize=10,rotation= 80)
sns.countplot(x=df['manufacturer'], hue = df['smartphone'], order=df.manufacturer.value_counts().sort_values
              (ascending = False).index)
plt.title('Manufacturer that sold the most')
```

```python
Samsung = df.loc[df.manufacturer == 'Samsung']
Samsung
```

```python
plt.figure(figsize = (10, 5))
plt.xticks(rotation = 'vertical')
sns.barplot(x='model', y ='units_sold_m', data = Samsung )
plt.title('Most sold model')
```
